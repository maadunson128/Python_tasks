```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.array([10,20,30,40,50])
y = np.array([13,24,56,77,23])
```

```python
def average(w1,w2):
    list = [ (w1 * x + w2 - y) **2 for xi,yi in zip(x,y)]
    return np.mean(list)
```

```python
w1_list = []
w2_list = []
err_list = [] 
```

```python
for w1 in range(11):
    for w2 in range(11):
        error = average(w1,w2)
        w1_list.append(w1)
        w2_list.append(w2)
        err_list.append(error)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = '3d')
ax.plot_trisurf(w1_list, w2_list, err_list, cmap = 'viridis')
plt.show()
```
